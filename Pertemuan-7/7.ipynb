```python
df_can_sorted= df_can.sort_values(by='Total', axis=0, ascending=False)
df_top5= df_can_sorted.head(5)
df_top5
```

```python
df_top5.shape
```

```python
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(28,16))

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Top 5 Countries Immigrating to Canada 1980-2013')
```

```python
df_top5.plot(kind='area', figsize=(14,8), stacked=False, alpha=0.3)

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Top 5 Countries Immigrating to Canada 1980-2013')
```

```python
area_plot_top5= df_top5.plot(kind='area', figsize=(14,8), stacked=False, alpha=0.3)

area_plot_top5.set_title('Imigrasi top 5 ke Canada 1980-2013')
area_plot_top5.set_ylabel('Jumlah Imigran')
area_plot_top5.set_xlabel('Tahun')
```
